## Marco Teórico: ANOVA en Profundidad

### Fundamento Matemático

ANOVA descompone la variabilidad total en:

$$SS_{Total} = SS_{Entre} + SS_{Dentro}$$

Donde:
- $SS_{Entre} = \sum_{j=1}^{k} n_j(\bar{Y}_j - \bar{Y})^2$ — variabilidad entre grupos
- $SS_{Dentro} = \sum_{j=1}^{k}\sum_{i=1}^{n_j}(Y_{ij} - \bar{Y}_j)^2$ — variabilidad dentro de grupos

El **estadístico F** es el cociente de las varianzas:
$$F = \frac{MS_{Entre}}{MS_{Dentro}} = \frac{SS_{Entre}/(k-1)}{SS_{Dentro}/(N-k)}$$

Un **F grande** indica que las diferencias entre grupos superan la variabilidad natural dentro de los grupos → evidencia contra H₀.

### Supuestos de ANOVA (Verificación Obligatoria)

| Supuesto | Prueba estadística | Umbral de decisión |
|----------|-------------------|-------------------|
| **Normalidad** de residuos | Shapiro-Wilk (n < 50) / Kolmogorov-Smirnov | p > 0.05 → normalidad aceptable |
| **Homocedasticidad** (varianzas iguales) | Levene / Bartlett | p > 0.05 → varianzas homogéneas |
| **Independencia** de observaciones | Diseño experimental | Verificar aleatorización |

**¿Qué hacer si se violan los supuestos?**
- Normalidad leve: ANOVA es robusto (Teorema del Límite Central con n ≥ 30)
- Sin normalidad con muestras pequeñas → Prueba Kruskal-Wallis (no paramétrica)
- Sin homocedasticidad → ANOVA de Welch (no asume varianzas iguales)

### Pruebas Post-Hoc

ANOVA solo indica **si existe** al menos una diferencia significativa, pero **no cuál par** de grupos difiere. Las pruebas post-hoc identifican los pares específicos, corrigiendo por múltiples comparaciones:

| Prueba | Corrección | Cuándo usar |
|--------|-----------|-------------|
| **Tukey HSD** | Controla tasa de error familiar | Todos los pares, grupos del mismo tamaño |
| **Bonferroni** | Divide α entre número de comparaciones | Pocos pares específicos de interés |
| **Games-Howell** | No asume varianzas iguales | Cuando Levene p < 0.05 |
| **Dunnett** | Comparar todos vs. control | Experimentos con grupo control |

### Tamaño de Efecto — η² (Eta-Cuadrado)

El p-valor indica **si hay** diferencia, pero no **cuán grande** es. El tamaño de efecto $\eta^2$ mide la magnitud:

$$\eta^2 = \frac{SS_{Entre}}{SS_{Total}}$$

| η² | Interpretación (Cohen) |
|----|----------------------|
| 0.01-0.06 | Efecto pequeño |
| 0.06-0.14 | Efecto mediano |
| > 0.14 | Efecto grande |

### Semáforo de Resultados ANOVA

| Resultado | Interpretación |
|-----------|----------------|
| p < 0.001 + η² > 0.14 | Diferencia grande y altamente significativa |
| p < 0.05 + η² > 0.06 | Diferencia mediana y significativa |
| p < 0.05 + η² < 0.06 | Diferencia significativa pero efecto pequeño |
| p > 0.05 | Sin evidencia de diferencia entre grupos |


## Caso de Estudio: Efecto de la Contaminación en la Salud de Arrecifes Coralinos

**Problema:** Científicos ambientales evalúan el impacto de diferentes niveles de contaminación (baja, media, alta) sobre la salud de arrecifes coralinos distribuidos en múltiples regiones. Quieren saber si los niveles de contaminación afectan significativamente el estado de salud.

**Técnica:** ANOVA de una vía (one-way ANOVA) compara las medias de salud del coral entre los tres grupos de contaminación.

**Hipótesis:**
- **H₀**: μ_baja = μ_media = μ_alta (las medias son iguales en los tres grupos)
- **H₁**: Al menos un par de medias difiere significativamente (p < 0.05)

### Diseño del Experimento

| Grupo | Nivel de Contaminación | n | Media Esperada | Desv. Est. |
|-------|----------------------|---|----------------|------------|
| 1 | Baja | 30 | 80 (alta salud) | 10 |
| 2 | Media | 30 | 60 (salud moderada) | 15 |
| 3 | Alta | 30 | 40 (salud deteriorada) | 20 |

### Flujo de Análisis

1. Generar datos sintéticos que simulan mediciones de campo
2. Verificar supuestos de normalidad y homocedasticidad
3. Aplicar ANOVA y evaluar significancia del estadístico F
4. Si ANOVA es significativo → prueba post-hoc Tukey HSD para identificar qué pares difieren
5. Calcular tamaño de efecto (η²) para evaluar la magnitud de las diferencias
6. Visualizar distribuciones con boxplots y violin plots

Case Use: Effect of Pollution on Coral Reef Health Across Different Regions
Problem Description:
Environmental scientists are concerned about the impact of varying levels of pollution on coral reef health. They have categorized several regions based on their pollution levels and want to understand how these levels affect coral health.

Statistical Technique:
Analysis of Variance (ANOVA) will be used to compare the means of coral health across different pollution level groups to determine if there are significant differences.

Hypothesis:
The hypothesis is that coral reef health significantly differs across regions with low, medium, and high pollution levels.

Simulation Setup:
For this simulation, we will generate synthetic data assuming measurements from 90 coral reef sites divided into three groups based on pollution levels:

Low Pollution (30 sites)
Medium Pollution (30 sites)
High Pollution (30 sites)
Coral health will be scored on a scale from 0 to 100, where 100 indicates optimal health.

```python
# ============================================================
# PASO 1: IMPORTACIONES Y CONFIGURACIÓN
# ============================================================
# numpy: álgebra numérica y generación de datos aleatorios
# pandas: manejo estructurado de datos
# scipy.stats: funciones estadísticas (f_oneway para ANOVA, shapiro para normalidad)
# matplotlib + seaborn: visualizaciones de calidad
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# Semilla para reproducibilidad — garantiza obtener siempre los mismos datos
np.random.seed(42)

# ============================================================
# PASO 2: GENERACIÓN DE DATOS SINTÉTICOS
# ============================================================
# Simulamos índices de salud del coral (escala 0-100) en 3 niveles de contaminación.
# La media disminuye conforme aumenta la contaminación (relación inversa esperada).
# La desviación estándar aumenta en grupos más contaminados (mayor variabilidad por estrés).
low_pollution_health    = np.random.normal(80, 10, 30)  # Contaminación baja: alta salud
medium_pollution_health = np.random.normal(60, 15, 30)  # Contaminación media: salud moderada
high_pollution_health   = np.random.normal(40, 20, 30)  # Contaminación alta: salud deteriorada

# Organizar en DataFrame: columna de salud + columna de grupo (factor categórico)
data = pd.DataFrame({
    'Coral Health': np.concatenate([low_pollution_health, medium_pollution_health, high_pollution_health]),
    'Pollution Level': ['Low']*30 + ['Medium']*30 + ['High']*30
})

# ============================================================
# PASO 3: ESTADÍSTICAS DESCRIPTIVAS POR GRUPO
# ============================================================
# Antes de ANOVA, siempre explorar los datos: ¿Las medias difieren visualmente?
# ¿La variabilidad es similar entre grupos (supuesto de homocedasticidad)?
print("Estadísticas Descriptivas por Nivel de Contaminación:")
print(data.groupby('Pollution Level')['Coral Health'].describe().round(2))
print()

# ============================================================
# PASO 4: ANOVA DE UNA VÍA
# ============================================================
# scipy.stats.f_oneway realiza ANOVA. Retorna el estadístico F y el p-valor.
# F grande → las medias entre grupos varían más que la variabilidad dentro de grupos.
# p < 0.05 → rechazar H₀ → al menos una media es diferente.
anova_result = stats.f_oneway(
    data[data['Pollution Level'] == 'Low']['Coral Health'],
    data[data['Pollution Level'] == 'Medium']['Coral Health'],
    data[data['Pollution Level'] == 'High']['Coral Health']
)

print(f"Resultado ANOVA:")
print(f"  F-estadístico: {anova_result.statistic:.4f}")
print(f"  p-valor: {anova_result.pvalue:.6f}")
if anova_result.pvalue < 0.05:
    print("  → RECHAZAR H₀: Hay diferencias significativas entre niveles de contaminación")
else:
    print("  → NO RECHAZAR H₀: Sin evidencia de diferencias significativas")

# ============================================================
# PASO 5: VISUALIZACIÓN
# ============================================================
# El boxplot muestra la distribución de cada grupo: mediana, cuartiles y outliers.
# Permite verificar visualmente si las medias difieren y si las varianzas son similares.
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(x='Pollution Level', y='Coral Health', data=data, palette='Set2',
            order=['Low', 'Medium', 'High'], ax=ax)
sns.stripplot(x='Pollution Level', y='Coral Health', data=data,
              order=['Low', 'Medium', 'High'], color='black', alpha=0.4, size=4, ax=ax)
ax.set_title('Salud del Coral por Nivel de Contaminación\nBoxplot + Puntos Individuales',
             fontweight='bold')
ax.set_xlabel('Nivel de Contaminación')
ax.set_ylabel('Índice de Salud del Coral')
ax.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# ANÁLISIS ANOVA EXTENDIDO CON VERIFICACIÓN DE SUPUESTOS Y TUKEY HSD
# ============================================================
# Esta versión incluye:
#   1. Verificación formal de normalidad (Shapiro-Wilk) y homocedasticidad (Levene)
#   2. ANOVA con decisión automática basada en supuestos
#   3. Prueba post-hoc Tukey HSD para comparaciones pares
#   4. Visualizaciones detalladas con violinplot + boxplot
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.multicomp import pairwise_tukeyhsd

np.random.seed(42)

# ============================================================
# PASO 1: DATOS CON ESCALA ECOLÓGICA (0-10 en lugar de 0-100)
# ============================================================
# Escala 0-10 más común en índices ecológicos compuestos.
# Misma estructura: 3 grupos con medias decrecientes y variabilidad controlada.
low_pollution_health    = np.random.normal(8, 1, 30)  # Arrecifes en buen estado
medium_pollution_health = np.random.normal(6, 1, 30)  # Arrecifes en estado moderado
high_pollution_health   = np.random.normal(4, 1, 30)  # Arrecifes degradados

data = pd.DataFrame({
    'Coral Health': np.concatenate([low_pollution_health, medium_pollution_health, high_pollution_health]),
    'Pollution Level': ['Low']*30 + ['Medium']*30 + ['High']*30
})

# ============================================================
# PASO 2: VERIFICACIÓN DE SUPUESTOS
# ============================================================
# SUPUESTO 1 — NORMALIDAD (Prueba Shapiro-Wilk)
# H₀: Los datos de cada grupo siguen distribución normal.
# Si p > 0.05 para todos los grupos → supuesto cumplido.
print("=" * 60)
print("VERIFICACIÓN DE SUPUESTOS ANOVA")
print("=" * 60)
print()
print("Prueba de Normalidad (Shapiro-Wilk):")
groups = {'Low': low_pollution_health, 'Medium': medium_pollution_health, 'High': high_pollution_health}
all_normal = True
for name, grp in groups.items():
    stat, p = stats.shapiro(grp)
    result = "✓ Normal" if p > 0.05 else "✗ No Normal"
    print(f"  {name:8s}: W={stat:.4f}, p={p:.4f}  → {result}")
    if p <= 0.05:
        all_normal = False

print()
# SUPUESTO 2 — HOMOCEDASTICIDAD (Prueba de Levene)
# H₀: Las varianzas de todos los grupos son iguales.
# Si p > 0.05 → varianzas homogéneas → supuesto cumplido.
levene_stat, levene_p = stats.levene(low_pollution_health, medium_pollution_health, high_pollution_health)
homoscedastic = levene_p > 0.05
print(f"Prueba de Homocedasticidad (Levene):")
print(f"  F={levene_stat:.4f}, p={levene_p:.4f}  → {'✓ Varianzas homogéneas' if homoscedastic else '✗ Varianzas heterogéneas'}")
print()

# ============================================================
# PASO 3: ANOVA SEGÚN SUPUESTOS
# ============================================================
# Si los supuestos se cumplen → ANOVA clásico (f_oneway)
# Si no se cumple homocedasticidad → ANOVA de Welch (más robusto)
print("=" * 60)
print("RESULTADO ANOVA")
print("=" * 60)
f_stat, p_val = stats.f_oneway(low_pollution_health, medium_pollution_health, high_pollution_health)
print(f"F-estadístico = {f_stat:.4f}")
print(f"p-valor       = {p_val:.6f}")
print(f"Significancia: {'RECHAZAR H₀ — diferencias significativas (p < 0.05)' if p_val < 0.05 else 'NO rechazar H₀'}")
print()

# ============================================================
# PASO 4: PRUEBA POST-HOC TUKEY HSD
# ============================================================
# Solo se realiza si ANOVA es significativo.
# Tukey HSD compara todos los pares de grupos y controla el error tipo I familiar.
# Ventaja sobre múltiples t-tests: mantiene α = 0.05 para el conjunto de comparaciones.
if p_val < 0.05:
    print("=" * 60)
    print("TUKEY HSD — COMPARACIONES PARES")
    print("=" * 60)
    tukey_result = pairwise_tukeyhsd(
        endog=data['Coral Health'],    # Variable dependiente
        groups=data['Pollution Level'], # Variable de agrupación (factor)
        alpha=0.05                      # Nivel de significancia
    )
    print(tukey_result)
    print()
    print("INTERPRETACIÓN: 'reject=True' indica que ese par de grupos")
    print("difiere significativamente en salud del coral (p-adj < 0.05).")

# ============================================================
# PASO 5: VISUALIZACIONES MÚLTIPLES
# ============================================================
# Panel de 4 gráficos: boxplot, violin, medias con IC, distribuciones
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
order_list = ['Low', 'Medium', 'High']
palette = 'Set2'

# 1. Boxplot con puntos individuales
sns.boxplot(x='Pollution Level', y='Coral Health', data=data,
            order=order_list, palette=palette, ax=axes[0,0])
sns.stripplot(x='Pollution Level', y='Coral Health', data=data,
              order=order_list, color='black', alpha=0.3, size=4, ax=axes[0,0])
axes[0,0].set_title('Distribución por Grupo (Boxplot + Datos)', fontweight='bold')
axes[0,0].set_ylabel('Índice de Salud del Coral')
axes[0,0].grid(True, alpha=0.3, axis='y')

# 2. Violin plot — muestra la forma completa de la distribución
sns.violinplot(x='Pollution Level', y='Coral Health', data=data,
               order=order_list, palette=palette, ax=axes[0,1])
axes[0,1].set_title('Distribución Completa (Violin Plot)', fontweight='bold')
axes[0,1].set_ylabel('Índice de Salud del Coral')
axes[0,1].grid(True, alpha=0.3, axis='y')

# 3. Medias con intervalos de confianza al 95%
means = data.groupby('Pollution Level')['Coral Health'].mean()
sems  = data.groupby('Pollution Level')['Coral Health'].sem()
for i, level in enumerate(order_list):
    axes[1,0].bar(i, means[level], yerr=sems[level]*1.96,
                  color=plt.cm.Set2(i/3), capsize=6, edgecolor='k', alpha=0.8)
axes[1,0].set_xticks(range(3))
axes[1,0].set_xticklabels(order_list)
axes[1,0].set_title('Medias con IC 95%', fontweight='bold')
axes[1,0].set_ylabel('Índice de Salud del Coral')
axes[1,0].grid(True, alpha=0.3, axis='y')

# 4. Texto de resultados del test
result_text = (
    f"ANOVA de Una Vía\n\n"
    f"F = {f_stat:.4f}\n"
    f"p-valor = {p_val:.6f}\n"
    f"Significativo: {'Sí' if p_val < 0.05 else 'No'} (α=0.05)\n\n"
    f"Supuestos:\n"
    f"  Normalidad: {'✓' if all_normal else '✗'}\n"
    f"  Homocedasticidad: {'✓' if homoscedastic else '✗'}"
)
axes[1,1].text(0.1, 0.5, result_text, transform=axes[1,1].transAxes,
               fontsize=12, verticalalignment='center',
               bbox=dict(boxstyle='round', facecolor='lightcyan', alpha=0.8))
axes[1,1].axis('off')
axes[1,1].set_title('Resumen de Resultados', fontweight='bold')

plt.tight_layout()
plt.show()
```
